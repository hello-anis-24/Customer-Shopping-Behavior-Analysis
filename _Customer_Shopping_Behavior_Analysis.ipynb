```python
# python code to insert big csv file into mysql database.
import mysql.connector
import math

# ---------------- LOAD CSV ----------------
df = pd.read_csv(
    "customer_shopping_behavior.csv",
    encoding="latin1"
)

# ---------------- CONNECT TO MYSQL ----------------
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="1234",
    database="customer_behavior"
,
    autocommit=False
)

cursor = conn.cursor()

# ---------------- INSERT QUERY ----------------
sql = """
INSERT INTO customer (
    customer_id,
    age,
    gender,
    item_purchased,
    category,
    purchase_amount_usd,
    location,
    size,
    color,
    season,
    review_rating,
    subscription_status,
    shipping_type,
    discount_applied,
    promo_code_used,
    previous_purchases,
    payment_method,
    frequency_of_purchases
)

VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s,%s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# ---------------- BULK INSERT WITH NaN FIX ----------------
batch_size = 10000
batch = []

for row in df.itertuples(index=False):
    cleaned_row = tuple(
        None if (isinstance(value, float) and math.isnan(value)) else value
        for value in row
    )
    batch.append(cleaned_row)

    if len(batch) == batch_size:
        cursor.executemany(sql, batch)
        conn.commit()
        batch.clear()

# Insert remaining rows
if batch:
    cursor.executemany(sql, batch)
    conn.commit()

# ---------------- CLOSE ----------------
cursor.close()
conn.close()

print(" Bulk insert completed successfully without NaN errors!")
```
